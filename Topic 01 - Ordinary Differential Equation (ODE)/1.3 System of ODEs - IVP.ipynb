## 1.3 System of ODEs - IVP

**Exercise 1:** Springy Pendulum

A mass $m$ is attached to a spring with spring constant $k$, which is attached to a support point as shown in the figure. The length of the resulting pendulum at any given time is the spring rest length $x_0$ plus the stretch (or compression) $x$, and the angle of the pendulum with respect to the vertical is $\theta$.

![alt text](images/springy-pendulum.png "Springy Pendulum")

The differential equations for this system are given by:
$$\ddot x = ({x_0} + x){\dot \theta ^2} - \frac{k}{m}x + g\cos \theta $$
$$\ddot \theta  =  - \frac{1}{{{x_0} + x}}\left[ {g\sin \theta  + 2\dot x\dot \theta } \right]$$

Let constants be

$x_0 = 1$ spring natural length

$k = 3.5$ spring constant

$m = 0.2$ mass in $kg$

$g = 9.8$ $m/s^2$

With these initial values:

$x(t=0) = 1$ initial spring extension

$\dot {x}(t=0) = 0$ initial velocity

$\theta(t=0) = 0.3$ initial angle in radians

$\dot {\theta}(t=0) = 0$ initial angular velocity

Write a program to plot the positions of the mass $m$ in $x-y$ coordinate over the time interval $0 \le t \le 25\; s$.

```python
from scipy.integrate import odeint
import matplotlib.pyplot as plt
import numpy as np
import math
```

```python
t = np.arange(0, 100.0, 0.01)

# define given constants
x_0 = 1
k = 3.5
m = 0.2
g = 9.8

# setup initial values vector [ x_dot_0, theta_dot_0, x_0, theta_0]
x_0 = 1
x_dot_0 = 0
theta_0 = 0.3
theta_dot_0 = 0

y_init = [x_dot_0, theta_dot_0, x_0, theta_0]


# define rhs function, f(y,t)
def rhs(y, t):
    return [
        (x_0 + y[2]) * (y[1]**2) - k/m*y[2] + g*math.cos(y[3]),
        -(g*math.sin(y[3]) + 2*y[0]*y[1])/(x_0 + y[2]),
        y[0],
        y[1]
    ]

y = odeint(rhs, y_init, t)

plt.plot(y[:,2], y[:,3])
# but it does not make sense to plot x vs theta
# the question wants x, y
```

```python
plt.plot(t, y[:,2])
```

```python
plt.plot(t, y[:,3])
```

```python
pos_x = (x_0 + y[:, 2])*np.sin(y[:, 3])
pos_y = -(x_0 + y[:, 2])*np.cos(y[:, 3])

plt.plot(pos_x, pos_y)
```
