$$u(0, t) = 2\sin(\omega t)$$

```python

c = 0.5
w = 10
maxx = 1 
maxt = 5
dt = 0.01
dx = 0.005
eps = ((dt*c/dx)**2)
print(eps)
'''
u(x,0) = 0 

row = j = t
col = i = pos

'''

n_col = int(maxx/ dx)
n_row = int(maxt/ dt)
print(n_col,n_row)
ts = np.linspace(0,maxt,n_row)
xs = np.linspace(0,maxx,n_col)
# print(ts)

#init 
u = np.zeros([n_col,n_row]) 
for row in range(1,len(u)):
    for col in range(1,len(u[0])-1):
        u[row][0] = 2*np.sin(w*0.01*row)
        u[1][col] = 0.5*eps * (u[0][col+1]+u[0][col-1]) + (1. - eps)* u[0][col]
        

#calculate col, t > 1
for row in range(1,len(u)-1):
    for col in range(1,len(u[0])-1):
        u[row+1][col] = eps * (u[row][col+1]+u[row][col-1]) + 2.*(1. - eps)* u[row][col] - u[row-1][col]
        
        
#end boundary        
for row in range(1,len(u)-1):
    u[row+1][-1] = -c *(u[row][-1]-u[row][-2]) * (dt/dx) - u[row][-1]
    print(f'{-c}*({u[row][-1]}-{u[row][-2]})*{(dt/dx)}-{u[row][-1]}')

print(len(u[0,:]),u)

print(u)
```

```python
fig, ax = plt.subplots()
ax = plt.axes(xlim=(0, 1), ylim=(-3, 3))
line, = ax.plot(ts,u[0,:])


def animate(i):
    line.set_data(ts,u[i,:])
    return line

anim = animation.FuncAnimation(fig, animate, frames=len(u[:,0]), repeat=False)
anim.save('wave-d1.gif', fps = 20)
plt.show
```

$$u(0, t) = 2\sin(\omega t)$$

and the boundary condition on the other end, $x = L$ is an **open boundary condition**.

$$u(x, 0) = 0$$
