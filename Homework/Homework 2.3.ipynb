### Homework 2.3 - 2D Temperature Distribution

Solve 2D heat equation for the following initial conditions:

$$\begin{array}{l}
u(x,0) = \sin (2\pi x)\\
u(x,1) = \sin (2\pi x)\\
u(0,y) = \sin (3\pi y)\\
u(1,y) = \sin (\pi y)
\end{array}$$

where $u(x, y)$ represent the steady state temperature distribution on 2D plane.
 
Plot the 2D heat map of the solution.

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.animation as animation
#%matplotlib inline
```

```python
maxx = 1
maxy = 1
dx = 0.01
dy = 0.01
Nx = maxx/dx
Ny = maxy/dy
u = np.zeros([int(Ny),int(Nx)])
h =0.00000002
             
#init Ux
for col in range(len(u[0])):
    u[0][col] = np.sin(2*np.pi*col*dx)
    u[-1][col] = np.sin(2*np.pi*col*dx)
    
# init Uy 
for row in range(len(u)):
    u[row][0] = np.sin(3*np.pi*row*dy)
    u[row][-1] = np.sin(np.pi*row*dy)

# run 
nloop = 0 
uold = u[1][1]
while True:       
    for row in range(1,len(u)-1):
        for col in range(1,len(u[0])-1):
            u[row][col] = 1/4*(u[row-1][col]+u[row+1][col]+u[row][col-1]+u[row][col+1])
    unew = u[1][1]
    print(unew,uold,abs(uold - unew))
    nloop += 1
    print(nloop)                
    if abs(uold - unew) < h:
        break 
    else:        
        uold = unew                        
                
    
plt.imshow(u.transpose(),cmap='hot')
```
